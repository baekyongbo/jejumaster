```python
import pandas as pd
import matplotlib.pyplot as plt
PATH = './sample_data/'
FILE = '거주국별_한국_여행_중_주요_방문지_20231101080547.csv'


file = PATH+FILE
df = pd.read_csv(file, encoding = 'cp949')
df
```

```python
# 컬럼 이름 변경하기(분류->국가)
df.columns = ['국가', '명동', '동대문시장',
              '고궁','남산','인사동','청계천',
              '이태원','신촌','강남역']
df.columns
```

```python
#국가 컬럼에서 아시아 5개국 데이터만 출력
asian_countries = ['일본', '중국', '태국', '인도', '말레이시아']
asian_data = df[df['국가'].isin(asian_countries)]
asian_data
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 코랩에서 한글 폰트 설치한 후 사용하기!!!
# plt.rc('font', family='cmb10')                  # (코랩)한글 폰트
plt.rc('font', family='NanumBarunGothic')       # (코랩)한글 폰트
# plt.rc('font', family='Malgun Gothic')          # (win)한글 폰트

# 아시아 5개 국가의 데이터 선택
asian_countries = ['일본', '중국', '태국', '말레이시아', '인도']
asian_data = df[df['국가'].isin(asian_countries)]

# 그래프 그리기
plt.figure(figsize=(12, 8))
x = asian_data['국가']
bar_width = 0.1
index = range(len(x))
colors = ['navy', 'mediumpurple', 'orange', 'teal', 'olive', 'lime', 'indigo', 'coral', 'plum']

for i, column in enumerate(asian_data.columns[1:]):
    plt.bar([pos + i * bar_width for pos in index], asian_data[column], width=bar_width, label=column, color=colors[i])

plt.title('아시아 5개 국가의 관광지 관심도', fontdict={'size': 16, 'color': 'black', 'weight': 'bold'}, pad=20)
plt.xlabel('국가', fontdict={'size': 12, 'color': 'dimgray', 'weight': 'bold'})
plt.ylabel('관심도', fontdict={'size': 12, 'color': 'dimgray', 'weight': 'bold'}, labelpad=20)
plt.xticks([pos + 2 * bar_width for pos in index], x)
plt.legend()
plt.grid(True, axis='y', alpha=0.5)

plt.show()
```
